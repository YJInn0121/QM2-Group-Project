```python
import pandas as pd
from statsmodels.formula.api import ols

new_df1['Life Satisfaction 2011/2012'] = pd.to_numeric(new_df1['Life Satisfaction 2011/2012'], errors='coerce')

formula = 'Q("Life Satisfaction 2011/2012") ~ Q("Anxiety 2011/2012") + Q("Gcse results 2011/2012") + Q("APS 2011/2012") + Q("Crime 2011/2012") + Q("income 2011/2012")'


model = ols(formula, data=new_df1).fit()

print(model.summary())


```
