```python
from google.colab import files
uploaded=files.upload()
! pip install pandas
import pandas as pd
df = pd.read_excel("gcse-results-old.xls")
df

from google.colab import files
uploaded = files.upload()


!pip install pandas plotly

import json
import pandas as pd
import plotly.express as px

geojson_filename = "borough_boundaries.json"

with open("borough_boundaries.json", "r") as file:
    geojson_data = json.load(file)



fig = px.choropleth_mapbox(df,
                           geojson=geojson_data,
                           locations='Unnamed: 1',
                           featureidkey="properties.name",
                           color="Percentage of pupils at the end of key stage 4 achieving 5+ A*-C including English & mathematics GCSEs",
                           color_continuous_scale="reds",
                           hover_name="Unnamed: 1",
                           mapbox_style="carto-positron",
                           center={"lat": 51.5072, "lon": 0.1276},
                           zoom=10,
                           opacity=0.5,
                           width=1100,
                           height=600,
                           )


fig.update_geos(fitbounds="locations", visible=False)

scatter_fig = px.scatter_geo(df,
                             lon='Longitude',
                             lat='Latitude',
                             hover_name="Unnamed: 1",
                             color="Percentage of pupils at the end of key stage 4 achieving 5+ A*-C including English & mathematics GCSEs",
                             color_continuous_scale="reds",
                             size_max=15,
                             size="Percentage of pupils at the end of key stage 4 achieving 5+ A*-C including English & mathematics GCSEs",
                             )
scatter_trace = scatter_fig['data'][0]
fig.add_trace(scatter_trace)

fig.show()
```
