```python
from google.colab import files
uploaded=files.upload()
! pip install pandas
import pandas as pd
df = pd.read_excel("residence earning.xlsx")
df.dropna(how='all', inplace=True)
pd.set_option('display.max_columns', None)
df.drop(df.index[0])
df.drop(df.index[34:52], inplace=True)
df.columns
clean = df.drop([2002,  'Unnamed: 3',
                2003,  'Unnamed: 5',          2004,  'Unnamed: 7',
                2005,  'Unnamed: 9',          2006, 'Unnamed: 11',
                2007, 'Unnamed: 13',          2008, 'Unnamed: 15',
                2009, 'Unnamed: 17',          2010, 'Unnamed: 19',
         'Unnamed: 21','Unnamed: 23','Unnamed: 25','Unnamed: 27','Unnamed: 29',
          2016, 'Unnamed: 31',
                2017, 'Unnamed: 33',          2018, 'Unnamed: 35',
                2019, 'Unnamed: 37',          2020, 'Unnamed: 39',
                2021, 'Unnamed: 41',          2022, 'Unnamed: 43'], axis=1)

for year in range(2011,2016):
  clean[year] = pd.to_numeric(clean[year], errors='coerce')
clean['income 2011/2012'] = clean[[2011, 2012]].mean(axis=1, numeric_only=True)
clean['income 2012/2013'] = clean[[2012, 2013]].mean(axis=1, numeric_only=True)
clean['income 2013/2014'] = clean[[2013, 2014]].mean(axis=1, numeric_only=True)
clean['income 2014/2015'] = clean[[2014, 2015]].mean(axis=1, numeric_only=True)
cleaned_income= clean.groupby('Area')['income 2011/2012','income 2012/2013','income 2013/2014','income 2014/2015'].mean().reset_index()
cleaned_income
```
